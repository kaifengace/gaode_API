### 地理/逆地理编码

```python
import requests
```

```python
#地理编码
key_hu = '28444c601f47ed94592282d899ec1b94'
def geo(address):
    """获取地理编码"""
    url = 'https://restapi.amap.com/v3/geocode/geo?parameters'
    params = {
        'key':key_hu,
        'address':address,
        'output':'json'
    }
    response = requests.get(url,params=params)
    data = response.json()
    return data
```

```python
geo('广东省深圳市龙岗区布吉')
```

```python
key_hu = '28444c601f47ed94592282d899ec1b94' # 高德地图API 获取Key
def geo(address:str,city:str)-> dict:
    """获取地理编码
    address可填入国家、省份、城市、区县、城镇、乡村、街道、门牌号码、屋邨、大厦等建筑物名称
    填入address对应city即可查询具体经纬度
    """
    parameters = {'key':key_hu,
                  'city':city,
                  'citylimit': True,
                  'address':address
                   } 
    r = requests.get ("https://restapi.amap.com/v3/geocode/geo?parameters", params=parameters)
    data = r.json()['geocodes'][0]['location']
    return data
```

```python
geo("广东省广州市从化区中山大学南方学院","广州")
```

```python
import requests
def poi(keywords:str,city:str)-> dict:
    """搜索POI
    输入keywords关键词
    输入搜索的范围城市city
    即可得到相关信息的表格"""
    parameters = {
        'key':key_hu,
        'city':city,
        'keywords':keywords,
        'output':'json',
        
    }
    r = requests.get ("https://restapi.amap.com/v3/place/text?parameters", params=parameters)
    data = r.json()
    pois = []
    pois.extend(data['pois'])
    import pandas as pd
    df_input = pd.DataFrame(pois)
    return df_input
```

### 静态地图

```python
def map(location:str,zoom:str):
    """zoom：设置地图大小
       location：选择地图中心坐标
    """
    parameters = {
        'key':key_hu,
        'location':location,
        'zoom':zoom,
        
    }
    r = requests.get ("https://restapi.amap.com/v3/staticmap?parameters", params=parameters)
    data = r.url
    return data
```

```python
map(geo("广东省广州市从化区中山大学南方学院","广州"),18)
```

### 坐标转换

```python
def assistant(locations:str,coordsys:str):
    """坐标转换
    locations填入非高德坐标
    coordsys可选值：gps,mapbar,baidu,autonavi（不进行转换）
    """
    parameters = {
        'key':key_hu,
        'coordsys':coordsys,
        'locations':locations,
        'output':'json'
    }
    r = requests.get ("https://restapi.amap.com/v3/assistant/coordinate/convert?parameters", params=parameters)
    data = r.json()['locations']
    return data
```

```python
assistant("116.481499,39.990475|116.481499,39.990375",'gps')
```

### 出行路线指引应用
#### 一、所掉用的高德API：
- 路径规划；
- 交通态势；
- 静态地图；
- 天气查询
#### 二、产品需求：
- 现今，导航类产品广泛的应用在这个社会，人们对这类产品的需求越来越高，因此，本产品为满足人们的需求制作了一款出行导航app，其中，功能有路径规划，交通态势还有静态地图。

#### 三、产品描述：
- 用户打开app输入目的地，显示出现路线方案，点开同时显示每条路径方案的交通态势，包括拥堵程度等，同时还有即时天气显示，提醒用户当前路径上的天气预报，提醒用户做好出行准备。
